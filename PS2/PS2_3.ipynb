## 3. Explore a data set

```python
import numpy as np
import pandas as pd
import seaborn as sns
from matplotlib import pyplot as plt
```

#### 3.1 Load the csv, XLS, or XLSX file, and clean possible data points with missing values or bad quality.
注明：本题目用到的数据集来自https://quotsoft.net/air/ ，为北京市2021年10月23日的空气质量数据

```python
# Read 2281305.csv as SZ_wind
Air_Qua = pd.read_csv('beijing_extra_20211023.csv',sep=',') #读入csv文件
Air_Qua.head(200)#预览数据
```

```python
#数据清洗，clean possible data points with missing values or bad quality
new_Air_Qua = Air_Qua.dropna() #使用dropna()删除原数据中有空值的行
new_Air_Qua.reset_index(drop=True, inplace=True)
new_Air_Qua.head(200)#预览结果
```

#### 3.2 Plot the time series of a certain variable
选取SO2在北京各地的统计平均浓度做时间序列图

```python
temp = new_Air_Qua[new_Air_Qua.columns[3:]] #截取有数据的列
new_Air_Qua['Average'] = temp.mean(axis=1) #对每个地点的数据求平均值,存入新的一列‘Average’中
new_Air_Qua.head(1000) #预览数据
```

```python
new_Air_Qua_SO2 =new_Air_Qua[new_Air_Qua['type'] == 'SO2']
new_Air_Qua_SO2 = new_Air_Qua_SO2.reset_index()
print(new_Air_Qua_SO2)
new_Air_Qua_SO2.plot(x='hour',y='Average') #绘制了10月23日各小时里SO2的统计平均浓度的时间序列
```

#### 3.3 Conduct at least 5 simple statistical checks with the variable, and report your findings.


```python
print('Mean of SO2 average is ',new_Air_Qua_SO2['Average'].mean())
print('Max of SO2 average is ',new_Air_Qua_SO2['Average'].max())
print('Min of SO2 average is ',new_Air_Qua_SO2['Average'].min())
print('Var of SO2 average is ',new_Air_Qua_SO2['Average'].var())
print('Quantile of SO2 average is ',new_Air_Qua_SO2['Average'].quantile())
```
